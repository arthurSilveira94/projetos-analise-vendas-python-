```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar dados
df = pd.read_csv('dataset_vendas.csv')

# Visualização inicial
print(df.head())

# Converter data
df['data'] = pd.to_datetime(df['data'])

# Criar coluna de faturamento
df['faturamento'] = df['quantidade'] * df['preco']

# Verificar dados nulos
print(df.isnull().sum())

# Análise 1 — Faturamento total
faturamento_total = df['faturamento'].sum()
print("Faturamento total:", faturamento_total)

# Análise 2 — Faturamento por produto
faturamento_produto = df.groupby('produto')['faturamento'].sum()
print(faturamento_produto)

# Análise 3 — Faturamento por região
faturamento_regiao = df.groupby('regiao')['faturamento'].sum()
print(faturamento_regiao)

# Análise 4 — Vendas ao longo do tempo
df.groupby('data')['faturamento'].sum().plot()
plt.title('Faturamento ao longo do tempo')
plt.xticks(rotation=45)
plt.show()

# Gráfico por produto
faturamento_produto.plot(kind='bar')
plt.title('Faturamento por Produto')
plt.show()

# Gráfico por região
faturamento_regiao.plot(kind='bar')
plt.title('Faturamento por Região')
plt.show()

# Insights:
# - Produto B teve maior faturamento
# - Região Sudeste concentra mais vendas
# - Há crescimento/queda em determinados períodos
```
